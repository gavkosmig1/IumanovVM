```python
import heapq

def prim(graph, start_vertex):
    vertices = len(graph)
    visited = [False] * vertices
    min_heap = [(0, start_vertex)]  # (вес, вершина)
    mst_weight = 0
    mst_edges = []
    
    while min_heap:
        current_weight, current_vertex = heapq.heappop(min_heap)
        
        if not visited[current_vertex]:
            visited[current_vertex] = True
            mst_weight += current_weight
            if current_weight != 0:
                mst_edges.append((current_vertex, prev_vertex, current_weight))
                
            for neighbor, weight in graph[current_vertex].items():
                if not visited[neighbor]:
                    heapq.heappush(min_heap, (weight, neighbor))
                    
            prev_vertex = current_vertex
    
    return mst_weight, mst_edges

# Пример использования
graph = {
    0: {1: 10, 2: 6, 3: 5},
    1: {0: 10, 3: 15},
    2: {0: 6, 3: 4},
    3: {0: 5, 1: 15, 2: 4}
}
start_vertex = 0

mst_weight, mst_edges = prim(graph, start_vertex)
print(f"Вес минимального остовного дерева: {mst_weight}")
print("Рёбра минимального остовного дерева:")
for edge in mst_edges:
    print(edge)
```

```python
edges = [
    (0, 1, 1),
    (0, 2, 2),
    (1, 2, 3),
    (1, 3, 4),
    (2, 3, 5),
    (3, 4, 20),
    (4, 5, 8),
    (4, 6, 11),
    (5, 6, 10),
    (6, 7, 1)
]
num_vertices = 8

class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.height = [0] * n

    def find_root(self, u) -> int:
        if self.parent[u] != u:
            self.parent[u] = self.find_root(self.parent[u])
        return self.parent[u]

    def merge(self, u, v) -> bool:
        root_u = self.find_root(u)
        root_v = self.find_root(v)
        
        if root_u != root_v:
            if self.height[root_u] > self.height[root_v]:
                self.parent[root_v] = root_u
            elif self.height[root_u] < self.height[root_v]:
                self.parent[root_u] = root_v
            else:
                self.parent[root_v] = root_u
                self.height[root_u] += 1
            return 1
        return 0


def kruskal(edges, num_vertices):
    uf = UnionFind(num_vertices)
    mst_edges = []
    
    edges.sort(key=lambda x: x[2])
    
    for u, v, weight in edges:
        if uf.find_root(u) != uf.find_root(v):
            uf.merge(u, v)
            mst_edges.append((u, v, weight))
            
    mst_edges.sort(key=lambda x: x[0])
    return mst_edges


mst = kruskal(edges, num_vertices)
print("Минимальное остовное дерево:", mst)
```
