```python
class Graph:
    def __init__(self, num_vertices):
        self.V = num_vertices
        self.graph = {}
    
    # Добавление ребра в граф
    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        
        self.graph[u].append(v)
        self.graph[v].append(u)

# Функция для выполнения DFS и проверки на наличие моста
def dfs(v, visited, parent, low, disc, time, bridges):
    global timer
    visited[v] = True
    disc[v] = low[v] = timer
    timer += 1
    
    for neighbor in graph.graph[v]:
        if not visited[neighbor]:
            parent[neighbor] = v
            dfs(neighbor, visited, parent, low, disc, time, bridges)
            
            low[v] = min(low[v], low[neighbor])
            
            # Если низкий индекс соседа больше времени открытия текущей вершины,
            # значит это мост
            if low[neighbor] > disc[v]:
                if (v > neighbor):
                    bridges.append((neighbor, v))
                else:
                    bridges.append((v, neighbor))
                
        elif neighbor != parent[v]:  # Обновляем low[v]
            low[v] = min(low[v], disc[neighbor])


data = input().split()

V = int(data[0])  # Количество вершин
E = int(data[1])  # Количество рёбер

edges = []

# Чтение рёбер
for _ in range(E):
    data = input().split()
    S = int(data[0])
    D = int(data[1])
    edges.append((S, D))

# Создаем граф
graph = Graph(V)

# Добавляем рёбра в граф
for edge in edges:
    graph.add_edge(edge[0], edge[1])

# Массив для хранения посещенных вершин
visited = [False] * V

# Массивы для хранения времени открытия и низкого индекса
disc = [-1] * V
low = [-1] * V

# Массив для хранения родителя каждой вершины
parent = [-1] * V

# Список для хранения мостов
bridges = []

# Глобальный счетчик времени
timer = 0

# Запускаем DFS из каждой непосещенной вершины
for i in range(V):
    if not visited[i]:
        dfs(i, visited, parent, low, disc, timer, bridges)

# Вывод результата
if len(bridges) == 0:
    print("Yes")
else:
    print(len(bridges))
    bridges.sort()
    for bridge in bridges:
        print(bridge[0], bridge[1])
```

```python
def bellman_ford(n, edges):
    dist = [float('inf')] * n
    dist[0] = 0  # Предполагаем, что стартовая вершина - 0

    for _ in range(n - 1):
        updated = False
        for u, v, w in edges:
            if dist[u] != float('inf') and dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                updated = True
        if not updated:
            break

    for u, v, w in edges:
        if dist[u] != float('inf') and dist[u] + w < dist[v]:
            return "Yes"
    return "No"

data = input().split()

V = int(data[0])  # Количество вершин
E = int(data[1])  # Количество рёбер

index = 2
edges = []

# Чтение рёбер
for _ in range(E):
    data = input().split()
    S = int(data[index])
    D = int(data[index + 1])
    W = int(data[index + 2])
    edges.append((S, D, W))

result = bellman_ford(V, edges)
print(result)
```
